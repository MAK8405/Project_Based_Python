```python
while True:
    user_guess = input("Guess a number between 1 and 100: ")

    if user_guess == "q":
        print("Thank you for playing. Goodbye!")
        break

    # Convert string to integer safely
    try:
        user_guess = int(user_guess)
    except ValueError:
        print("Invalid input. Please enter a number.")
        continue

    # Range check
    if user_guess > 100 or user_guess < 1:
        print("Out of range! Pick a number between 1 and 100.")
        continue

    print(user_guess)

# We are writing our code based on Objective
```

```python
def validate_input(user_guess):
    if not user_guess.isdigit():
        print("Invalid input. Please try again.")
        return False

    user_guess = int(user_guess)
    if user_guess > 100 or user_guess < 1:
        print(
            "Your guess is out of range. Please try again. Your guess should be between 1 and 100."
        )
        return False

    return True
```

```python
score = 100

while True:
    user_guess = input("Guess a number between 1 and 100: ")
    if user_guess == "q":
        print("Thank you for playing. Goodbye!")
        break
    if not validate_input(user_guess):
        continue
    # Convert to integer
    user_guess = int(user_guess)

    if rand_num > user_guess:
        print("Your guess is too low. Please try again.")
    elif rand_num < user_guess:
        print("Your guess is too high. Please try again.")
    else:
        print("Congratulation! You guessed the correct number!")
        print(f"Your score is: {score}")
        break
    score -= 10
    score = max(score, 0)
```
