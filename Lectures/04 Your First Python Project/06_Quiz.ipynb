**5Ô∏è‚É£ How can you see the version of an installed Python package (e.g., numpy)? (Select all that apply)**

a) Use the command pip show numpy  
b) Use the command pip list | grep numpy  
c) Use the command pip freeze | grep numpy  
d) Use the command pip list --format=freeze | grep numpy  

---

<details>
  <summary>‚úÖ Show Answer</summary>

**Answer:** a), b), c), d)  

üí° **Explanation:**  
All the given choices are correct.  

- **Option 1:** This command will display detailed information about the specified package, including its version.  
- **Option 2:** The `pip list` command lists all installed packages, and the `grep` command is used to search for a specific package name within the output. In this case, it will display the package name and version for numpy if it is installed.  
- **Option 3:** The `pip freeze` command generates a list of installed packages and their versions in a format suitable for creating a `requirements.txt` file. By piping the output to `grep`, you can filter for the specific package name, numpy, and view its version.  
- **Option 4:** This command is similar to option 3, but the `--format=freeze` flag explicitly sets the output format to the freeze format. Again, piping the output to `grep` allows you to filter for the package name, numpy, and see its version.  

üí° **Hint:** When using option 3 to view the version of a package and generate a `requirements.txt` file, you might encounter a situation where the output displays something like `@ file://...` instead of the actual package version. This behavior can occur when you have installed a package using a local file path instead of the PyPI package index. When `pip freeze` encounters such a package, it includes the file path instead of a version number in the output. To address this issue, you can use option 4.  

</details>
