تا موقعی که پروژه تموم نشده نیاز نیست ریدمی دات امدی رو بنویسی.

وقتی میخوای کد بزنی توی پایتون دوتا سلوشن پیش روی ما هست برای تست و دیباگینگ یکی فایل دات پای و یکی جوپیتور نوت بوک هستش.

فایل های جوپیتر برای تست کردن خیلی خوبن چون در واقع repl اند

read
evaluation
print
loop
معمولا اول با این شروع میکنیم و بعد سلوشن رو میبریم توی فایل دات پای که یک نفر بتونه از بالا تا پایین این اسکریپت رو ران بکنه و فلدر های مختلف و کارای مختلف.
